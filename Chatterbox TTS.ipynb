## ðŸ–¥ï¸ Check GPU Availability

First, let's verify that a GPU is available and check its specifications.

````python
# Check if GPU is available
import torch
import psutil
import platform
import GPUtil

print("=" * 50)
print("ðŸ” System Resource Check")
print("=" * 50)

# Check CUDA availability
if torch.cuda.is_available():
    print("âœ… CUDA is available")
    
    # Get GPU information
    gpus = GPUtil.getGPUs()
    if gpus:
        gpu = gpus[0]
        print(f"ðŸŽ® GPU: {gpu.name}")
        print(f"ðŸ’¾ GPU Memory: {gpu.memoryTotal:.2f} GB")
        print(f"ðŸ“Š GPU Memory Free: {gpu.memoryFree:.2f} GB")
        
        if gpu.memoryTotal >= 5:
            print("âœ… GPU has sufficient memory (â‰¥5GB)")
        else:
            print("âš ï¸ Warning: GPU has less than 5GB of memory")
    else:
        print("âš ï¸ Warning: No GPU detected")
else:
    print("âŒ CUDA is not available. Please enable GPU in Runtime settings.")

# Check CPU and RAM
cpu_count = psutil.cpu_count()
memory = psutil.virtual_memory()
print(f"ðŸ–¥ï¸ CPU: {platform.processor()} ({cpu_count} cores)")
print(f"ðŸ’¾ RAM: {memory.total / (1024**3):.2f} GB")
print("=" * 50)
````
